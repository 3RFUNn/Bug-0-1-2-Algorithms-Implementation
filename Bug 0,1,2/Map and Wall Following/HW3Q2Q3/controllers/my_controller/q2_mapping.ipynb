```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
# Path to your .txt file
file_path = 'map_data.txt'

# Lists to store the Cartesian coordinates
x_values = []
y_values = []
xs_robot = []
ys_robot = []
distance = 0.05

# Read the file
with open(file_path, 'r') as file:
    for line in file:
        x, y, compass0, compass1 = line.split(',')

        x_robot = -float(x.strip())
        y_robot = -float(y.strip())

        compass0 = float(compass0.strip())
        compass1 = float(compass1.strip())

        teta = math.atan2(compass0, compass1)
        sin_teta = np.sin(teta)
        cos_teta = np.cos(teta)

        x_wall = x_robot + sin_teta * distance
        y_wall = y_robot - cos_teta * distance

        x_values.append(x_wall)
        y_values.append(y_wall)
        xs_robot.append(x_robot)
        ys_robot.append(y_robot)

# Create a scatter plot

plt.scatter(x_values, y_values,c='r')
plt.scatter(xs_robot, ys_robot,c='silver')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Map')
plt.legend(['map','robot path'],loc='upper left')
plt.show()
```
